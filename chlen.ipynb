```python
def drop_by_iqr(col_name, data):
    iqr = data[col_name].quantile(.75) - data[col_name].quantile(.25)
    median = data[col_name].median()
    return data[(-1.5 * iqr <= data[col_name] - median) & (data[col_name] - median <= 1.5 * iqr)]
```

```python
df = drop_by_iqr('Аналитический блок', df)
```

```python
histdf['Класс'].value_counts()
```

```python
from scipy.stats import mannwhitneyu
sample_1 = df[df['Класс'] == '11']['Аналитический блок']
sample_2 = df[df['Класс'] == '10']['Аналитический блок']
mannwhitneyu(sample_1, sample_2).pvalue
from scipy.stats import ttest_ind

ttest_ind(sample_1, sample_2, equal_var=False)
```
